###### Department wise codekata performence (pie chart)

```python
df.groupby(["Department"])["CodeKata Score"].sum()
```

```python
w = list(df["Department"].unique())
plt.figure(figsize = (8,8))
explode = [0.05, 0.05, 0.04]
colors = ['cyan','grey','orange']
plt.pie(df.groupby(["Department"])["CodeKata Score"].sum(), labels=w, colors=colors, autopct='%1.1f%%', radius=0.8, shadow=False, data=True, explode=explode )
plt.title('---- Department-wise CODEKATA Performance ----')
plt.legend(bbox_to_anchor=(-0.6,0.6), loc='upper left')

# Displays Department wise codekata performence in form of "" ___PIE Chart*___"
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
df.Department.unique()

# Displays "all departments name" which is present inside the df 
```

```python
EEE = df[df['Department']=='Electronics and Electrical Engineering'].sort_values('CodeKata Score', ascending=False).head(3)
CSE = df[df['Department']=='Computer Science and Engineering'].sort_values('CodeKata Score', ascending=False).head(3)
ECE = df[df['Department']=='Electronics and Communication Engineering'].sort_values('CodeKata Score', ascending=False).head(3)

highest = pd.concat([EEE, CSE, ECE], axis=0, ignore_index = True)
highest

# EEE -  Displays the EEE department topper details in Descending-order
# CSE -  Displays the CSE department topper details in Descending-order
# ECE -  Displays the ECE department topper details in Descending-order

# hhighest - Represent the concatination of all 3 department toppers
```

```python
fig = px.bar(
    highest,
    x="Department",
    y=["Name", "CodeKata Score"],
    color='Name',
    barmode='group',
    title='Department wise Toppers'    
)

fig.update_layout(
    height=600,
    width=800,
    font_size=16,
    font_family ='Arial',
    legend_title_font_color='blue',
    title= dict(
        x=0.1,
        y=0.95,
        font= {
               "family":'Time New Roman',
               "size":28,
               "color":'black'
              }
    )
)
fig.show()


# Displays Department wise Topper in form of "___Bar-chart*___"
```
